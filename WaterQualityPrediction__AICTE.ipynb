```python
import pandas as pd
import matplotlib.pyplot as plt

pollutants = ['O2', 'NO3', 'NO2', 'SO4','PO4', 'CL']

avg_concentrations = df[pollutants].mean()


plt.figure(figsize=(8, 8))
plt.pie(avg_concentrations, labels=pollutants, autopct='%1.1f%%', startangle=140, colors=plt.cm.Set3.colors)
plt.title('Average Concentration of Water Pollutants')
plt.axis('equal') 
plt.show()
```

```python
# Convert date column to datetime
df['date'] = pd.to_datetime(df['date'], dayfirst=True)

# Plot pollutants over time
df.set_index('date')[pollutants].plot(figsize=(14, 6))
plt.title('Pollutant Levels Over Time')
plt.ylabel('Concentration')
plt.xlabel('Date')
plt.legend(loc='upper right')
plt.show()
```

```python
df.set_index('date')[pollutants].plot(kind='area', stacked=True, figsize=(14, 6), alpha=0.6)
plt.title('Stacked Area Chart of Pollutants Over Time')
plt.ylabel('Concentration')
plt.xlabel('Date')
plt.show()
```

```python
df = df.dropna(subset=pollutants)
df.head()
```

```python
df.isnull().sum()
```

```python
X = df[['id', 'year']]
y = df[pollutants]
```

```python
X_encoded = pd.get_dummies(X, columns=['id'], drop_first=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)
```

```python
model = MultiOutputRegressor(RandomForestRegressor(n_estimators=100, random_state=42))
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("Model Performance on the Test Data:")
for i, pollutant in enumerate(pollutants):
    print(f'{pollutant}:')
    print('   MSE:', mean_squared_error(y_test.iloc[:, i], y_pred[:, i]))
    print('   R2:', r2_score(y_test.iloc[:, i], y_pred[:, i]))
    print()
```

```python
station_id = '22'
year_input = 2024

input_data = pd.DataFrame({'year': [year_input], 'id': [station_id]})
input_encoded = pd.get_dummies(input_data, columns=['id'])

# Align with training feature columns
missing_cols = set(X_encoded.columns) - set(input_encoded.columns)
for col in missing_cols:
    input_encoded[col] = 0
input_encoded = input_encoded[X_encoded.columns]  # reorder columns

# Predict pollutants
predicted_pollutants = model.predict(input_encoded)[0]

print(f"\nPredicted pollutant levels for station '{station_id}' in {year_input}:")
for p, val in zip(pollutants, predicted_pollutants):
    print(f"  {p}: {val:.2f}")
```
